#  Section B: Coding Questions


```python
 # 1. Check positive, negative, or zero

# Take a number input from the user
num = float(input("Enter a number: "))

# Check if the number is positive
if num > 0:
    print("The number is Positive")

# Check if the number is negative
elif num < 0:
    print("The number is Negative")

# Otherwise the number is zero
else:
    print("The number is Zero")

```

```python
# 2. Function to return sum and average

# Define a function to calculate sum and average
def calculate(numbers):
    # Calculate total sum
    total = sum(numbers)

    # Calculate average
    average = total / len(numbers)

    # Return both values
    return total, average


# Sample list of numbers
nums = [10, 20, 30, 40]

# Call the function
result = calculate(nums)

# Display results
print("Sum is:", result[0])
print("Average is:", result[1])
```

```python
# 3. Print first n even numbers

# Take input from the user
n = int(input("Enter how many even numbers you want: "))

# Loop from 1 to n
for i in range(1, n + 1):
    # Even number formula
    print(2 * i)
```

```python
# 4. Palindrome without built-in reverse

# Take string input from user
text = input("Enter a string: ")

# Initialize empty string for reversed text
reversed_text = ""

# Reverse the string manually
for char in text:
    reversed_text = char + reversed_text

# Compare original and reversed string
if text == reversed_text:
    print("It is a Palindrome")
else:
    print("It is not a Palindrome")
```

```python
# 5. NumPy array operations

# Import NumPy library
import numpy as np

# Create NumPy array
arr = np.array([10, 20, 30, 40, 50])

# Print the array
print("Array is:", arr)

# Print maximum value
print("Maximum value is:", np.max(arr))

# Print minimum value
print("Minimum value is:", np.min(arr))

# Print sum of elements
print("Sum of elements is:", np.sum(arr))
```

```python
# 6. Star pattern

# Outer loop for rows
for i in range(5):

    # Inner loop for columns
    for j in range(5):
        print("*", end=" ")

    # Move to next line after each row
    print()
```

```python
# 7. Alternate elements using slicing

# Create a list
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

# Get alternate elements using slicing
new_list = numbers[::2]

# Print original list
print("Original list:", numbers)

# Print sliced list
print("List with alternate elements:", new_list)
```
