```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load Excel dataset
df = pd.read_excel("4876484-Product_Sales_Dataset.xlsx")

print("Dataset Loaded Successfully ✅")
print("Shape of dataset:", df.shape)

df.head()
```

## Data Inspection

We inspect:
- Data structure
- Data types
- Missing values

```python
# Dataset shape
print("Dataset Shape:", df.shape)

# Data types and info
df.info()

# Summary statistics
df.describe()
```

```python
# Check null values
df.isnull().sum()
```

## Data Cleaning

Handling missing values and correcting data types if required.

```python
# Fill missing numerical values with mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Confirm no missing values
df.isnull().sum()
```

# Exploratory Data Analysis (EDA)

```python
# Calculate total sales
total_sales = df["Revenue"].sum()
print("Total Sales:", total_sales)
```

```python
# Calculate average sales
average_sales = df["Revenue"].mean()
print("Average Sales:", average_sales)
```

```python
# Group by Category
category_sales = df.groupby("Category")["Revenue"].sum().sort_values(ascending=False)

category_sales
```

## Sales Trend Over Time

```python
plt.figure(figsize=(10,5))
df.groupby("Order_Date")["Revenue"].sum().plot()
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

## Category-wise Sales Comparison

```python
plt.figure(figsize=(8,5))
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title("Category-wise Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

##  Transaction Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Category", data=df)
plt.title("Number of Transactions per Category")
plt.show()
```

# Key Findings & Business Insights

1. The highest revenue-generating category is identified.
2. Sales trends show peak and low performance periods.
3. Average sales value gives insight into customer purchase behavior.
4. Categories with high transactions but low revenue may need pricing optimization.
5. Business can focus marketing efforts on top-performing categories.
