Project Title:
Student Marks Analyzer Using Python and NumPy

Problem Statement:

The objective of this project is to develop a Student Marks Analyzer using Python and NumPy. The program accepts marks of multiple students, stores them in a NumPy array, and calculates total marks, average marks, highest marks, and lowest marks. Based on the average marks, the program assigns a grade using conditional statements.

Logic Explanation:

The program takes marks of multiple students using a loop and stores them in a list, which is converted into a NumPy array.

It calculates total, average, highest, and lowest marks using NumPy functions inside user-defined functions.

Based on the average marks, a grade is assigned using conditional statements (if-elif-else).

```python
# Student Marks Analyzer
import numpy as np

# User-Defined Functions
def calculate_total(marks_array):
    return np.sum(marks_array)

def calculate_average(marks_array):
    return np.mean(marks_array)

def find_highest(marks_array):
    return np.max(marks_array)

def find_lowest(marks_array):
    return np.min(marks_array)

def assign_grade(average):
    if average > 85:
        return "A"
    elif average >= 70:
        return "B"
    elif average >= 50:
        return "C"
    else:
        return "Fail"


print("===== STUDENT MARKS ANALYZER =====")

number_of_students = int(input("Enter number of students: "))

marks_list = []

# Taking marks input using loop
for i in range(1, number_of_students + 1):
    mark = float(input(f"Enter marks for Student {i}: "))
    marks_list.append(mark)

# Converting list to NumPy array
marks_array = np.array(marks_list)

# Calculations
total_marks = calculate_total(marks_array)
average_marks = calculate_average(marks_array)
highest_marks = find_highest(marks_array)
lowest_marks = find_lowest(marks_array)

grade = assign_grade(average_marks)

# Displaying Results
print("\n===== MARKS SUMMARY =====")
print("Total Marks:", total_marks)
print("Average Marks:", average_marks)
print("Highest Marks:", highest_marks)
print("Lowest Marks:", lowest_marks)
print("Assigned Grade:", grade)
```
