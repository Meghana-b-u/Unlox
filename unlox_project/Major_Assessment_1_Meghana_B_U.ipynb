2. Reverse a string

```python
def reverse_string(s):
  reversed_str=" "
  for char in s:
    reversed_str=char+reversed_str
  return reversed_str

text = input("Enter a string: ")
reversed_text = reverse_string(text)
print("Reversed string:", reversed_text)
```

3. Palindrome Number Function

```python
def is_palindrome(number):
    original = str(number)
    reversed_num = ""

    for char in original:
        reversed_num = char + reversed_num

    if original == reversed_num:
        return "Palindrome"
    else:
        return "Not a Palindrome"

num = int(input("Enter a number: "))
print(is_palindrome(num))
```

# **Section C**

6. Data Loading and Inspection

```python
import pandas as pd

#load dataset
df = pd.read_excel("filpcart.xlsx")

#first 5 rows
df.head()

#shape
print("dataset shape:" , df.shape)

#column names
print("columns:" , df.columns)

#basic info
df.info()
```

7. Data Cleaning

```python
# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
df = df.dropna()

# Convert price and rating to numeric
df["retail_price"] = pd.to_numeric(df["retail_price"], errors='coerce')
df["product_rating"] = pd.to_numeric(df["product_rating"], errors='coerce')

# Drop rows where conversion failed
df = df.dropna(subset=["retail_price", "product_rating"])

df.info()
```

8. Data Analysis

```python
# Top 5 categories by Average Rating
top_categories = df.groupby("product_category_tree")["product_rating"].mean().sort_values(ascending=False).head(5)
print("Top 5 categories by Average Rating:")
print(top_categories)

# 5 most expensive Products
expensive_products = df.nlargest(5, "retail_price")
print("\n5 most expensive Products:")
print(expensive_products[["product_name","retail_price"]])
```

9. Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Distribution of product prices

```python
plt.figure()
plt.hist(df["retail_price"], bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution of Product Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

Count plot of product categories

```python
plt.figure()
sns.countplot(data=df, x="product_category_tree")
plt.xticks(rotation=90)
plt.title("Count of Product Categories")
plt.show()
```

10. Insights

1) Certain product categories have significantly higher average rating, indication better customer satisfaction in those segments.

2) The price distribution is right_skewed, meaning most products are moderately priced, while only a few are very expensive.

3) Some categories dominate the dataset in terms of product count,showing higher product availability in those segments.
